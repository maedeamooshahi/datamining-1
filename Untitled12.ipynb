```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
groceries = pd.read_csv('/content/groceries1.csv')
```

```python
groceries.head(10)
```

```python
groceries.dtypes
```

```python
groceries.A.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))

plt.title(" histogram")

plt.ylabel('Number')

plt.xlabel('');
```

```python
df = pd.DataFrame()
clmns=[]
for i in list(range(0,32)):
    clmns = clmns+groceries[i].tolist()
clmns=list(np.unique(clmns))
dummies=pd.get_dummies(groceries)
for i in clmns:
    df[i] = dummies[[c for c in dummies.columns if i in c]].sum(axis=1)
```
